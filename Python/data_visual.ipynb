```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style(style = 'whitegrid')
plt.rcParams['figure.figsize'] = (8, 5)

df = pd.read_csv('netflix_customer_churn.csv')
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
#UNIVARIATE ANALYSIS
```

```python
#univariate analysis of age distribution
sns.histplot(df['age'], kde = True, bins = 20, color = 'skyblue')
plt.title('Distribution of Netflix Users')
plt.show()
```

```python
#categorical count plot
sns.countplot(x='gender', data=df, palette='Set2')
plt.title('Gender Distribution of Netflix Users')
plt.show()
```

```python
#Boxplot outlier detection
sns.boxplot(x='subscription_type', y='watch_hours', data=df, palette='cool')
plt.title('Watch Hours by Subscription Type')
plt.show()
```

```python
#voilin plot for watch hours by subscription type
sns.violinplot(x='subscription_type', y='watch_hours', data=df, palette='muted')
plt.title('Watch Hours Distribution by Subscription Type')
plt.show()
```

```python
#BIVARIATE ANALYSIS
```

```python
#Scatter plot (continuous vs continuous)
sns.scatterplot(x = 'watch_hours', y = 'monthly_fee', hue = 'churned', data = df, alpha = 0.7)
plt.title('Watch Hours vs Monthly Fee by Churn Status')
plt.show()
```

```python
#correlation heatmap
corr = df[['age', 'watch_hours', 'monthly_fee', 'churned','last_login_days',
            'avg_watch_time_per_day','number_of_profiles']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#barplot (categorical vs numerical)
sns.barplot(x = 'region', y = 'churned' , data=df , estimator = lambda x: sum(x)/len(x))
plt.title('Churn Rate by Region')
plt.ylabel('Churn Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
#MULTIVARIATE ANALYSIS
```

```python
#pairplot (relationships across features)
sns.pairplot(df[['age', 'watch_hours', 'monthly_fee', 'churned']], hue='churned')
plt.show()
```

```python
#facet grid (multiple small plots)
g = sns.FacetGrid(df, col='region', hue='churned', height=3, col_wrap=3)
g.map(sns.scatterplot, 'watch_hours', 'avg_watch_time_per_day', alpha=0.7)
g.add_legend()
plt.show()
```

```python
#heatmap by category
pivot = df.pivot_table(values='churned', index='subscription_type', columns = 'region', aggfunc='mean')
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.title('Churn Rate by Subscription Type and Region')
plt.show()
```

```python
#asthetic customization and storytelling
plt.figure(figsize=(10,6))
sns.barplot(x='favorite_genre', y='watch_hours', hue='churned', data=df, palette='Spectral')
plt.title('Average Watch Hours by Favorite Genre and Churn Status')
plt.xlabel('Favorite Genre')
plt.xticks(rotation=45)
plt.ylabel('Average Watch Hours')
plt.legend(title='Churned', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
#bonus: interactive visualization with plotly
import plotly.express as px
fig = px.scatter(df, x='watch_hours', y='monthly_fee', color='subscription_type',
                 size='avg_watch_time_per_day',
                 hover_data=['region', 'churned'],
                 title='Watch Hours vs Monthly Fee by Subscription Type')
fig.show()
```
