```python
import pandas as pd
import numpy as np

# 1. Let's make a new dataset where an HR employee's age is missing
company_data = {
    'Name': ['Sam', 'John', 'Sara', 'Emily', 'Alex', 'Kate'],
    'Department': ['Tech', 'Tech', 'Tech', 'HR', 'HR', 'HR'],
    'Age': [22, 24, 26, np.nan, 40, 42]  # Emily's age in HR is missing!
}
df_company = pd.DataFrame(company_data)

print("--- Data Before Smart Imputation ---")
print(df_company)

# 2. Group by Department, calculate the median age for THAT department, and fill the missing value
df_company['Age'] = df_company['Age'].fillna(df_company.groupby('Department')['Age'].transform('median'))

print("\n--- Data After Smart Imputation ---")
print(df_company)
```

```python
# 1. Add a temporary text priority column to practice label encoding
df_company['Priority'] = ['Low', 'Medium', 'High', 'Medium', 'Low', 'High']

print("--- Data With Text Priority Column ---")
print(df_company)

# 2. Define your map
priority_map = {'Low': 1, 'Medium': 2, 'High': 3}

# 3. Safely map the text to numbers
df_company['Priority_num'] = df_company['Priority'].map(priority_map)

print("\n--- Data After Mapping Priority ---")
print(df_company)
```

```python
# Convert the 'Department' text column into numerical columns
# drop_first=True helps prevent redundant data (if it's not Tech, it must be HR!)
df_encoded = pd.get_dummies(df_company, columns=['Department'], drop_first=True, dtype=int)

print("--- One-Hot Encoded DataFrame ---")
print(df_encoded)
```
