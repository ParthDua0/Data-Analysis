```python
import pandas as pd
import numpy as np

# 1. THE MESSY DATASET
raw_data = {
    'Player': ['Alpha', 'Beta', 'Gamma', 'Gamma', 'Delta', 'Epsilon', 'Zeta', 'Eta', 'Theta', 'Iota'],
    'Division': ['Masters', 'Amateur', 'Masters', 'Masters', 'Amateur', 'Masters', 'Amateur', 'Amateur', 'Masters', 'Amateur'],
    'Match_Rating': ['1850 pts', '1100 pts', np.nan, np.nan, '950 pts', '1900 pts', '2500 pts', '1050 pts', '1750 pts', '1200 pts'], # Zeta is a massive outlier!
    'Tier': ['Gold', 'Bronze', 'Silver', 'Silver', 'Bronze', 'Gold', 'Bronze', 'Bronze', 'Silver', 'Silver']
}
df = pd.DataFrame(raw_data)
print("=== 🟡 Raw Dataset ===")
print(df)
# ----------------------------------------------------
# 🛠️ YOUR CODE CHALLENGE STARTS HERE
# ----------------------------------------------------

# Task 1: Drop Duplicates
df = df.drop_duplicates()

# Task 2: Clean 'Match_Rating' string text and cast to float
df['Match_Rating'] = df['Match_Rating'].str.replace(' pts', '').astype(float)

# Task 3: Fill missing values using the median of their specific 'Division' group
df['Match_Rating'] = df['Match_Rating'].fillna(df.groupby('Division')['Match_Rating'].transform('median'))

# Task 4: Map 'Tier' text to numbers ('Bronze': 1, 'Silver': 2, 'Gold': 3)
tier_map = {'Bronze': 1, 'Silver': 2, 'Gold': 3}
df['Tier_num'] = df['Tier'].map(tier_map)

# Task 5: IQR Outlier Removal for 'Match_Rating'
Q1 = df['Match_Rating'].quantile(0.25)
Q3 = df['Match_Rating'].quantile(0.75)
IQR = Q3 - Q1
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

# Filter the dataframe using your fences
df_clean = df[(df['Match_Rating'] >= lower_fence) & (df['Match_Rating'] <= upper_fence)]

# Task 6: Group by 'Tier_num' and calculate the mean 'Match_Rating'
final_summary = df_clean.groupby('Tier_num')['Match_Rating'].mean()

# ----------------------------------------------------
# 📊 DISPLAY RESULTS
# ----------------------------------------------------
print("=== 🟢 Cleaned Dataset ===")
print(df_clean)
print("=== 🟢 Final Processed Summary ===")
print(final_summary)
```
